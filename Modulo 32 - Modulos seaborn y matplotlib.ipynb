### Histogramas

Representación gráfica de una variable en barras, donde la base de cada barra representa la frecuencia de la variable representada

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import seaborn as sns
```

```python
%matplotlib inline  
```

```python
datos1 = np.random.randn(100)
datos1
```

```python
#Hacemos un histograma de esos datos
```

```python
mpl.pyplot.hist(datos1)
```

```python
sns.distplot(datos1)
```

```python
sns.distplot(datos1,color='green')
```

```python
mpl.pyplot.hist(datos1,color='yellow', alpha = 0.2) #Color selecciona el color, alpha selecciona el grado de oscuridad
```

```python
#Vamos a mezclar dos histogramas
```

```python
datos2 = np.random.rand(80)
mpl.pyplot.hist(datos2, color = 'grey')
```

```python
mpl.pyplot.hist(datos1, color = 'green', alpha = 0.4, bins =20, density = True)
mpl.pyplot.hist(datos2, color = 'grey', alpha = 0.3, bins = 20, density = True)
# Ponemos "density = True" porque cada datos tiene distinta cantidad de registros, para que los "trate por igual"
```

### Cominamos estilos

```python
datos = np.random.randn(100)
```

```python
sns.distplot(datos, bins=10, color = 'green')
```

```python
sns.distplot(datos, color = 'green', rug = False, hist = False) #Así nos quedamos solo con la curva, sin las barras
```

```python
argumentos_curva = {'color':'black','label':'Curva'}
argumentos_histograma = {'color':'red','label':'Histograma'}
```

```python
sns.distplot(datos, bins=25,kde_kws=argumentos_curva, hist_kws = argumentos_histograma, label= True)
# kde_kws: para configurar los parámetros de la curva
# hist_kws: para configurar los parámetros de las barras
```

```python
serie = pd.Series(datos)
```

```python
sns.distplot(serie, bins=25, color = 'green')
```

### Regresiones Lineales

Para comprender como varia una variable en función del valor de otras variables

```python
propinas = sns.load_dataset('tips')
```

```python
propinas.head()
```

```python
#Vamos a crear un gráfico de regresión lineal con estos datos
```

```python
sns.lmplot('total_bill','tip',propinas) #Escribimos el valor del eje X, el eje Y, y el set de datos
```

```python
sns.lmplot('total_bill','tip',propinas, scatter_kws = {'marker':'*','color':'green'}, line_kws = {'color':'black'})
```

```python
sns.lmplot('total_bill','tip',propinas,fit_reg = False) #Con fit_reg = False, quitamos la línea de regresión
```

```python
propinas.head()
```

```python
#Vamos a crear una nueva propina: el porecntaje de la propina sobre el total
```

```python
propinas['porcentaje'] = 100 * propinas['tip']/propinas['total_bill']
```

```python
propinas.head()
```

```python
sns.lmplot('size','porcentaje',propinas)
```

```python
sns.lmplot('total_bill','porcentaje',propinas, hue='sex', markers =['x','o']) 
#Compara el total de la factura con el porcentaje, pero además lo hace también según el sexo
```

```python
sns.lmplot('total_bill','porcentaje',propinas, hue='day') 
#Compara el total de la factura con el porcentaje, pero además lo hace también según el día de la semana
```

### Mapas de calor (Heat map)

```python
 vuelos = sns.load_dataset('flights')
```

```python
vuelos.head()
```

```python
# Para hacer un mapa de calor necesitamos que sea una matriz
```

```python
vuelos = vuelos.pivot('month','year','passengers') #Filas, #Columnas, #Correspondencia
```

```python
vuelos
```

```python
#Ahora representamos esta misma matriz como un mapa de calor
```

```python
sns.heatmap(vuelos)
```

```python
sns.heatmap(vuelos, annot = True, fmt = 'd')
```

```python
vuelos.loc['May'][1956]
```

```python
#Vamos a tomar ese valor como valor centro, como color de referencia 
```

```python
valor_central = vuelos.loc['May'][1956]
valor_central
```

```python
sns.heatmap(vuelos, center = valor_central, annot = True, fmt = 'd')
```

```python
#Podemos cambiar la barra que indica el color abajo:
```

```python
sns.heatmap(vuelos, center = valor_central, annot = True, fmt = 'd', cbar_kws = {'orientation':'horizontal'})
```
